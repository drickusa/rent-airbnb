```python
# Read table from metastore / Unity Catalog
table_name = "rentalproperty.silver.airbnb"

# Folder to write parquet files into
# Using Unity Catalog Volume (serverless-compatible storage)
output_folder = "/mnt/rentalproperty/bronze/exports/silver_airbnb"

df = spark.table(table_name)

# Write parquet files
(
    df.write
      .mode("overwrite")          # use "append" if you want to add new files
      .format("parquet")
      .option("compression", "snappy")
      .save(output_folder)
)
```
