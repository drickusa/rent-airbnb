```python
%sql
/*when removing all NULL zipcodes and doing a distinct on the dataset we can use see that latitude and longitude can be used as the Primary Key of the silver table
 When checking on zip code it gives more than one property per zip code
*/
with distinct_airbnb 
    (select Distinct *
     from  rentalproperty.bronze.airbnb
     where zipcode is not null)
Select count(1),
       longitude,
       latitude
From distinct_airbnb
Group By longitude, latitude
Having Count(1) > 1 
```

```python
%sql
with distinct_airbnb 
    (Select * from rentalproperty.bronze.airbnb)
Select Distinct room_type
From distinct_airbnb
```

```python
%sql
with distinct_airbnb 
    (Select * from rentalproperty.bronze.airbnb)
Select *
From distinct_airbnb
Where review_scores_value is null
```

```python
%sql
--create PK using MD5 hash of latitude and longitude, not recommended to use type double for joining etc.. 
select Distinct md5(concat(Cast(longitude as string), cast(latitude as string))) as airbnb_pk,  * 
from  rentalproperty.bronze.airbnb
where zipcode is not null
```
